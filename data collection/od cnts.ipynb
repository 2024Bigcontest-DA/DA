## 데이터 조인

```python
data = pd.read_csv('data/data.csv')
data = data.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0'])
data
```

```python
transit = pd.read_csv('transit_columns.csv')
transit
```

```python
transit['출발지'] = transit.구분.str[0]
transit['목적지'] = transit.구분.str[1:]
transit
```

```python
hdong_cd = {3020055000: 'D',
            2635052000: 'B',
            5115057200: 'GA',
            5115058000: 'GB',
            4575034000: 'I',
            1156054000: 'S'
            }
data['목적지'] = data.dest_hdong_cd.map(hdong_cd)
data
```

```python
data.origin_cd.value_counts()
```

```python
code = {26: 'B',
        11: 'S',
        51: 'G',
        30: 'D',
        45: 'I'}
data['출발지'] = data.origin_cd.map(code)
data
```

```python
data = data[['출발지', '목적지', 'start_time', 'gender', 'age', 'origin_purpose', 'dest_purpose', 'od_dist_avg', 'od_duration_avg', 'od_cnts', 'modal']]
data
```

```python
transit['출발지'] = transit.구분.str[0]
transit['목적지'] = transit.구분.str[1:]
transit = transit[['출발지', '목적지', 'min_time', 'max_time', 'num_trainsit_methods', 'median_time', '예상 배차시간', '최대 이동거리', '최소 이동거리', '이동거리 중간값', '최대 교통수단 수', '환승횟수', '마지막 도보시간 중간값', '마지막 도보시간 평균']]
transit
```

```python
tot = pd.merge(data, transit, how = 'left', on = ['출발지', '목적지'])
tot
```

```python
tot.to_csv('tot_data_2.csv', index = False)
```
